```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# thin tailed (normal) distribution
trials = 30000
sum_to_now = 0
average_outcome_by_trial = []
for i in range(trials):
    draw = np.random.normal(0, 1)
    sum_to_now += draw
    average_outcome_by_trial.append(sum_to_now/(i+1))
```

```python
plt.plot(average_outcome_by_trial)
plt.show()
```

```python
# Now the fat tailed Cauchy distribution
trials = 30000
sum_to_now = 0
average_outcome_by_trial = []
for i in range(trials):
    draw = np.random.standard_cauchy()
    sum_to_now += draw
    average_outcome_by_trial.append(sum_to_now/(i+1))
```

```python
plt.plot(average_outcome_by_trial)
plt.show()
```
